```python
sns.countplot(data=data,x="Sex")
plt.title("Distribution by gender")
plt.show()
```

```python
sns.histplot(data=data,x="Age",bins=20,kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(data=data,x="Height",bins=20,kde=True)
plt.title("Height Distribution")
plt.show()
```

```python
sns.histplot(data=data,x="Weight",bins=20,kde=True)
plt.title("Weight Distribution")
plt.show()
```

```python
sns.countplot(data=data,x="Medal")
plt.title("Medal Distribution")
plt.show()
```

```python
sns.countplot(data=data,x="Year",hue="Medal")
plt.title("Medal Distribution Over the Years")
plt.show()
```

```python
year_avg_age=data.groupby("Year")["Age"].mean()
print(year_avg_age)
```

```python
sport_medium_height=data.groupby("Sport")["Height"].median()
print(sport_medium_height.max())
```

```python
sport_medium_height[sport_medium_height==190.0]
```

```python
country_gender_count=data.groupby(["NOC","Sex"])["ID"].count()
print(country_gender_count)
```

```python
country_gold_medals=data[data["Medal"]=="Gold"].groupby("NOC")["Medal"].count()
print(country_gold_medals.max())
```

```python
sport_gender_avg_wt=data.groupby(["Sport","Sex"])["Weight"].mean()
print(sport_gender_avg_wt["Wrestling"]["F"])
```

```python
sport_event_count=data.groupby("Sport")["Event"].nunique().sort_values(ascending=False)
sport_event_count.head(15).plot(kind="bar")
plt.title("Number of Unique Event per Sport")
plt.xlabel("Sport")
plt.ylabel("Number od Unique Events")
plt.show()
```

```python
year_partipition_count=data.groupby("Year")["ID"].nunique().sort_values()
year_partipition_count.plot(kind="bar")
plt.title("Number of Participants Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Participants")
plt.show()
```

```python
country_avg_age=data.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age.head(20).plot(kind="bar")
plt.title("Top 10 countries with highest Average Age of Participants")
plt.xlabel("Country")
plt.ylabel("Average Age")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=data,x="Season",y="Age")
plt.title("Distribution of Ages by Season")
plt.xlabel("Season")
plt.ylabel("Age")

plt.show()
```

```python
sns.violinplot(data=data,x="Medal",y="Height",palette="Set2")
plt.title("Distribution of Height by medal")
plt.xlabel("Medal")
plt.ylabel("Height")
plt.show()
```

```python
most_medals_country=data["NOC"].value_counts().idxmax()
print("Most medal-Wining Country: ",most_medals_country)
```

```python
tallest_athlete=data[data["Height"]==data["Height"].max()]
print("tallest_Athlete")
print(tallest_athlete[["ID","Name","Height","Sport"]])
```

```python
Heviest_athlete=data[data["Weight"]==data["Weight"].max()]
print("Heaviest_Athlete")
print(Heviest_athlete[["ID","Name","Weight","Sport"]])
```

```python
sns.scatterplot(data=data,x="Height",y="Weight",hue="Medal",palette="Set1")
plt.title("Athlete Height vs Weight by Medal Status")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend(title="Medal")
plt.show()
```

```python
medal_by_country_year=data.pivot_table(index="NOC",columns="Year",values="Medal",aggfunc="count")
sns.heatmap(medal_by_country_year,linewidths=0.5)
plt.title("Medal Counts By Country and Year")
plt.xlabel("Year")
plt.ylabel("Country")
plt.xticks(rotation=45)
plt.show()
```
